```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import zscore, norm
import matplotlib.pyplot as plt

def generate_1d_data(n_samples=300, n_anomalies=10, cluster_std=0.60, random_state=42):
    """Generate synthetic 1D data with some anomalies."""
    # Create clustered data in 1D
    np.random.seed(random_state)
    X = np.random.normal(loc=0, scale=cluster_std, size=n_samples)
    # Add random anomalies
    anomalies = np.random.uniform(low=-4, high=4, size=n_anomalies)
    # Combine normal data and anomalies
    return np.concatenate([X, anomalies])

def detect_anomalies_zscore(X, threshold=3):
    """Detect anomalies using Z-score method."""
    # Compute the Z-scores for the data
    z_scores = np.abs(zscore(X))  # No need to reshape here
    # Identify points where Z-score exceeds the threshold
    anomalies = X[z_scores > threshold]
    return anomalies, z_scores

def plot_1d_anomalies_with_distribution(X, anomalies, title="Anomaly Detection using Z-Score"):
    """Plot the 1D dataset with anomalies highlighted and Normal Distribution curve."""
    # Plot normal points
    normal_points = X[~np.isin(X, anomalies)]
    
    # Plot Normal Distribution curve
    mean = np.mean(X)
    std = np.std(X)
    
    # Generate x values for the Normal distribution curve
    x_vals = np.linspace(min(X)-1, max(X)+1, 1000)
    pdf_vals = norm.pdf(x_vals, mean, std)
    
    # Plot using Plotly
    fig = go.Figure()

    # Add normal points
    fig.add_trace(go.Scatter(
        x=normal_points, y=np.zeros_like(normal_points),
        mode='markers',
        marker=dict(size=8, color='blue', opacity=0.8),
        name='Normal Points'
    ))
    
    # Add anomalies
    fig.add_trace(go.Scatter(
        x=anomalies, y=np.zeros_like(anomalies),
        mode='markers',
        marker=dict(size=10, color='red', symbol='x', opacity=0.9),
        name='Anomalies'
    ))

    # Add the Normal Distribution curve
    fig.add_trace(go.Scatter(
        x=x_vals, y=pdf_vals,
        mode='lines',
        line=dict(color='green', width=2),
        name='Normal Distribution Curve'
    ))
    
    fig.update_layout(
        title=title,
        width=800,
        height=400,
        xaxis_title='Feature',
        yaxis=dict(
            title='Density',
            tickvals=[],
            showgrid=False,
            zeroline=False
        ),
        showlegend=True
    )
    fig.show()

def main():
    # Parameters
    n_samples = 300
    n_anomalies = 10
    cluster_std = 0.60
    z_threshold = 3  # Z-score threshold for anomaly detection
    
    # Generate and plot data
    X = generate_1d_data(n_samples=n_samples, n_anomalies=n_anomalies, cluster_std=cluster_std)
    
    # Detect anomalies using Z-score
    anomalies, z_scores = detect_anomalies_zscore(X, threshold=z_threshold)
    print(f"Number of detected anomalies: {len(anomalies)}")
    
    # Plot anomalies with Normal Distribution curve
    plot_1d_anomalies_with_distribution(X, anomalies, title=f"Anomaly Detection using Z-Score (Threshold = {z_threshold})")

if __name__ == "__main__":
    main()
```
