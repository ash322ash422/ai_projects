```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of students
n_students = 100

# Generate features with realistic relationships to learning style
study_hours = np.random.uniform(5, 20, n_students)  # Random study hours between 5 and 20 hours/week
class_participation = np.random.uniform(0, 10, n_students)  # Random participation score
previous_grades = np.random.uniform(60, 90, n_students)  # Random grades between 60 and 90
time_spent_on_homework = np.random.uniform(2, 10, n_students)  # Homework time in hours/week
motivation_level = np.random.uniform(1, 5, n_students)  # Motivation level on a scale of 1 to 5
teacher_feedback = np.random.uniform(1, 5, n_students)  # Teacher feedback on a scale of 1 to 5
learning_environment = np.random.choice([0, 1], n_students)  # 0 for home, 1 for library/study group

# Create learning_style based on realistic conditions
learning_style = []

# Visual learners may prefer more study hours, a higher level of class participation, or different feedback from teachers.
# Auditory learners might respond better to different environments, like lectures or audio content, and might have distinct feedback patterns.
# Kinesthetic learners may prefer hands-on activities, have specific teacher feedback related to engagement, and may need more time for practical application.
for i in range(n_students):
    # Visual learners
    if study_hours[i] > 15 and teacher_feedback[i] > 3:
        learning_style.append('visual')
    # Auditory learners
    elif class_participation[i] > 7 and motivation_level[i] > 3:
        learning_style.append('auditory')
    # Kinesthetic learners
    elif time_spent_on_homework[i] > 6 and motivation_level[i] < 3:
        learning_style.append('kinesthetic')
    else:
        # Default to 'visual' in other ambiguous cases
        learning_style.append('visual')

# Convert to DataFrame
df = pd.DataFrame({
    'student_id': range(1, n_students + 1),
    'learning_style': learning_style,
    'study_hours': study_hours,
    'class_participation': class_participation,
    'previous_grades': previous_grades,
    'time_spent_on_homework': time_spent_on_homework,
    'motivation_level': motivation_level,
    'teacher_feedback': teacher_feedback,
    'learning_environment': learning_environment
})

# Print out first few rows of the dataset to check
print(df.head())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split


# Prepare the data
X = df[['study_hours', 'class_participation', 'previous_grades', 'time_spent_on_homework', 
        'motivation_level', 'teacher_feedback', 'learning_environment']]
y = df['learning_style']

# Encode the target variable (learning_style)
y_encoded = pd.get_dummies(y)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train the model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Evaluate the model
score = model.score(X_test, y_test)
print(f'Model accuracy: {score:.2f}')
```

```python
# Sample student data for prediction
new_student = {
    'study_hours': 3,
    'class_participation': 2,
    'previous_grades': 3.1,
    'time_spent_on_homework': 6.1,
    'motivation_level': 2.23,
    'teacher_feedback': 3.73,
    'learning_environment': 0
}

new_student_df = pd.DataFrame([new_student])

# Predict the learning style
predicted_style = model.predict(new_student_df)
predicted_style_label = y_encoded.columns[predicted_style.argmax()]

# Recommend content based on the predicted learning style
content_recommendations = {
'visual': 'Watch this video tutorial on algebra.',
'auditory': 'Listen to this podcast on algebra concepts.',
'kinesthetic': 'Try this hands-on algebra activity.'
}

recommended_content = content_recommendations[predicted_style_label]
print(f'Recommended content: {recommended_content}')
```
